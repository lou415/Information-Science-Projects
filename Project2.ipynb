```python
# What do we want to do here?
# We want to group the districts by the classification
filtered_complaints = complaints_cleaned_2022[["district_occurrence", "general_cap_classification"]]
filtered_complaints
```

```python
# how many complaints are there in each district?
final_filtered_2022 = filtered_complaints.groupby("district_occurrence").value_counts().reset_index()
final_filtered_2022
```

```python
alt.Chart(final_filtered_2022).mark_bar().encode(
    x = alt.X("district_occurrence:N", title = "District Number", axis = alt.Axis(labelAngle = 0)),
    y = alt.Y("count", title = "Number of Complaints"),
    color = "general_cap_classification:N",
    tooltip = [alt.Tooltip("general_cap_classification:N", title="Complaint Type"), "count:Q"]
).properties(
    title = "Complaint Type and Count by District",
    width = 500
)
```

```python
# Complaints that stick out the most.
# Lack of service, Departmental violations, Unprofessional conduct, Physical abuse, Verbal Abuse, Civil Rights Complaint. 
```

```python
top_7 = ["LACK OF SERVICE", "DEPARTMENTAL VIOLATIONS", "UNPROFESSIONAL CONDUCT", "PHYSICAL ABUSE"]
```

```python
top_7_data = final_filtered_2022[final_filtered_2022["general_cap_classification"].isin(top_7)]
top_7_data
```

```python
top_7_chart = alt.Chart(top_7_data, title = alt.Title("Complaint Type and Count by District", subtitle = "Lack of service is among the most reported complaints")).mark_bar().encode(
    x = alt.X("district_occurrence:N", title = "District Number", axis = alt.Axis(labelAngle = 0)),
    y = alt.Y("count", title = "Number of Complaints"),
    color = alt.Color("general_cap_classification:N",legend = alt.Legend(orient="bottom", direction = "horizontal")),
    tooltip = [alt.Tooltip("general_cap_classification:N", title="Complaint Type"), "count:Q"],
).properties(
    width = 500
)

top_7_chart
```

```python
top_7_chart = alt.Chart(top_7_data).mark_bar().encode(
    x = alt.X("district_occurrence:N", title = "District Number", axis = alt.Axis(labelAngle = 0), sort=alt.EncodingSortField(field="count", order="descending")),
    y = alt.Y("count", title = "Number of Complaints"),
    color = alt.Color("general_cap_classification:N",legend = alt.Legend(orient="bottom", direction = "horizontal")),
    tooltip = [alt.Tooltip("general_cap_classification:N", title="Complaint Type"), "count:Q"],
).properties(
    width = 500
)
top_7_chart
```
